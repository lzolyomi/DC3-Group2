```python
def add_dash(string, index):
    #adds a dash to a string in position index
    return string[:index] + "-" + string[index:]

def format_datestring(string):
    #accepts a datestring from the rain timeseries and formats it
    date_only = add_dash(string, 8).split("-")[0] #removes the hour
    date_only = add_dash(date_only, 6) #separate day and month
    date_only = add_dash(date_only, 4) #separates month and year
    return date_only
```

```python
from os import listdir
from datetime import datetime
path = "/Users/levente/Documents/Quartile 1/Data Challenge 3/DC3-Group2/data/rain_historic_timeseries/"
files = listdir(path)
rain_file_dict = {}
for filename in files:
    split = filename.split('_')
    begin_date = format_datestring(split[3])
    begin_format = datetime.strptime(begin_date, "%Y-%m-%d")
    end_date = format_datestring(split[4].split('.')[0]) #get rid of csv
    end_format = datetime.strptime(end_date, "%Y-%m-%d")
    duration = abs((end_format - begin_format).days)
    rain_file_dict[begin_date] = filename
```

```python
from bs4 import BeautifulSoup
import requests

url = "https://veghelsweer.nl/noaa2/maandoverzicht.php?fichier=juli-2021"

response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")
```

```python
rows = soup.find("table")
data_rows = rows.find_all("tr")[2:] 
data_dct = {"Day":[], "Temp-min":[], "Temp-max":[], "Temp-avg":[], "Wind":[], "Rain":[]}
len(data_dct.keys())
```

```python
import numpy as np

data = np.array(data_rows[0].text.split("\n"))
indices = [1,2,3,4,5,7]
len(list(data[indices]))
data_rows[-1]
```

```python
indices = [1,2,3,4,5,7]
for row in data_rows: 
    formatted = np.array(row.text.split("\n"))
    if len(formatted) > 1:
        datapoints = list(formatted[indices])
        for pt, key in zip(datapoints,  data_dct.keys()):
            data_dct[key].append(pt)
pd.DataFrame(data_dct)
    
```
