```python
inProj = Proj('epsg:28992')
outProj = Proj('epsg:4326')
df["lat"] = 0.0
df["long"] = 0.0
for i in range(len(df)):
    x1= df["X"][i]
    y1 = df["Y"][i]
    x2, y2 = transform(inProj,outProj,x1,y1)
    df["lat"][i] = x2
    df["long"][i] = y2  
df
```
